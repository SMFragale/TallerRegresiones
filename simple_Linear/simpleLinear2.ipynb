```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('../datasets/diamonds.csv')
dataset.head(1)
```

```python
#Nombre de las variables a analizar (mismo nombre que la tabla dataset)
name_x = "depth"
name_y = "price"
data = dataset.loc[:,[name_x, name_y]]
```

```python
data.plot(x=name_x, y=name_y, style='o')
plt.xlabel(name_x)
plt.ylabel(name_y)
plt.show()
```

```python
X = pd.DataFrame(data[name_x])
Y = pd.DataFrame(data[name_y])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
regress = LinearRegression()
regress.fit(X_train, Y_train)
```

```python
y_pred = regress.predict(X_test)
```

```python
plt.scatter(X_train, Y_train, color = 'red')
plt.plot(X_train, regress.predict(X_train), color = 'blue')
plt.title(f'{name_y} vs {name_x} (Training set)')
plt.xlabel(name_x)
plt.ylabel(name_y)
plt.show()
```

```python
plt.scatter(X_test, Y_test, color = 'red')
plt.plot(X_test, regress.predict(X_test), color = 'blue')
plt.title(f'{name_y} vs {name_x} (Test set)')
plt.xlabel(name_x)
plt.ylabel(name_y)
plt.show()
```

```python

from sklearn.metrics import r2_score
print(regress.score(X_train, Y_train))
print(regress.coef_)
```
